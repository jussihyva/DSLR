```python
fname = os.path.join("../data", "dataset_train.csv")
my_data = pd.read_csv(fname)
```

```python
hogwarts_subjects = my_data[["Arithmancy", "Astronomy", "Herbology", "Defense Against the Dark Arts", "Divination"]]
mean = hogwarts_subjects.mean()
std = hogwarts_subjects.std()
median = hogwarts_subjects.median()
count = hogwarts_subjects.count()
max = hogwarts_subjects.max()
min = hogwarts_subjects.min()
q25 = hogwarts_subjects.quantile(0.25)
q50 = hogwarts_subjects.quantile(0.50)
q75 = hogwarts_subjects.quantile(0.75)
my_describe_df = pd.DataFrame({"Mean":mean, "Median":median, "Std":std, "Count":count, "Max":max, "Min":min, "25%":q25, "50%":q50, "75%":q75})
print(my_describe_df)
```
